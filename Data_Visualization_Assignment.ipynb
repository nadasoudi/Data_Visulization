```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define colors for "Absent" and "Present" categories
absent_color = '#FF69B4'  # Green color for "Absent"
present_color = '#6495ED'  # Red color for "Present"

# Setting seaborn aesthetics
sns.set(style='whitegrid', rc={"figure.figsize": (6, 4), "axes.titlesize": 18,
                               "axes.labelsize": 16, "xtick.labelsize": 14,
                               "ytick.labelsize": 16})

# Create countplot
a = sns.countplot(x='target', data=df, palette=[absent_color, present_color])

# Set title and axis labels
a.set_title('Distribution of Male vs. Female')

# Add titles to the bars
for i, p in enumerate(a.patches):
    height = p.get_height()
    if height > 0:
        # Calculate position for title
        title_y = p.get_y() + height / 2
        title_x = p.get_x() + p.get_width() / 2
        if i == 0:
            title_text = 'Female'
        else:
            title_text = 'Male'
        a.text(title_x, title_y, title_text, ha='center', va='center', color='black', fontsize=16)  # Increased fontsize
        a.annotate(format(height, '.0f'), (p.get_x() + p.get_width() / 2., p.get_y() -12+ height / 2),
                   ha='center', va='top', fontsize=20, color='black')

# Set x-axis tick labels
a.set_xticklabels(['', ''])

# Remove the y-axis label for cleanliness
a.set_ylabel('')

# Remove spines (borders) from the plot
sns.despine()

# Remove gridlines
a.grid(False)

# Remove x-axis label
a.set_xlabel('')
a.set_ylabel('')

# Set x-axis tick labels
#c.set_xticklabels(['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'], fontsize=14)
#a.set_yticklabels('')

# Show the plot
plt.xlabel("Distribution of Male vs. Female")
plt.show()
```

```python

# Calculate age categories
df['age_category'] = pd.cut(df.age, bins=list(np.arange(25, 85, 5)))

# Setting seaborn aesthetics
sns.set(style='whitegrid', palette='colorblind', rc={"figure.figsize": (16, 6),
                                                     "axes.titlesize": 18,
                                                     "axes.labelsize": 16,
                                                     "xtick.labelsize": 14,
                                                     "ytick.labelsize": 14})

# Create subplots
plt.subplot(121)
present_counts = df[df.target == 1].groupby('age_category')['age'].count()
present_counts.plot(kind='bar', width=0.8, color='#B85656')  # Set color for presence of heart disease
plt.title('Diseased')
plt.xlabel('Age Group')
plt.ylabel('Population Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Remove gridlines
plt.grid(False)

# Create subplots
plt.subplot(122)
absent_counts = df[df.target == 0].groupby('age_category')['age'].count()
absent_counts.plot(kind='bar', width=0.8, color='#56B86F')  # Set color for absence of heart disease
plt.title('Healthy')
plt.xlabel('Age Group')
plt.ylabel('Population Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Remove gridlines
plt.grid(False)
sns.despine()
# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

**EX5:Observing Correlations with a Heatmap**

**After Editing**

```python
# Create heatmap using the numeric DataFrame
f, ax = plt.subplots(figsize=(8, 8))
df_numeric = df.select_dtypes(include=['number'])
heatmap = sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', linewidths=0.5, fmt='.1f', ax=ax)

# Set title
heatmap.set_title('Correlation Heatmap')

# Adjust font sizes
heatmap.tick_params(axis='both', which='major', labelsize=12)
heatmap.tick_params(axis='both', which='minor', labelsize=10)
heatmap.set_xlabel('Features', fontsize=14)
heatmap.set_ylabel('Features', fontsize=14)

# Rotate y-axis labels for better readability
plt.yticks(rotation=0)

# Add gridlines
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
